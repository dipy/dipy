```python
import nibabel as nib

import numpy as np

import matplotlib.pylab as pl
import matplotlib.pyplot as plt

from dipy.io import read_bvals_bvecs
from dipy.core.gradients import gradient_table
from dipy.core.sphere import disperse_charges, Sphere, HemiSphere
from ivim import IvimModel, ivim_prediction
from dipy.io import read_bvals_bvecs

%matplotlib inline
```

```python
# Get bvecs for a set of bvals
def get_bvecs(n_pts):
    theta = np.pi * np.random.rand(n_pts)
    phi = 2 * np.pi * np.random.rand(n_pts)
    hsph_initial = HemiSphere(theta=theta, phi=phi)
    hsph_updated, potential = disperse_charges(hsph_initial, 5000)
    vertices = hsph_updated.vertices
    return vertices
```

```python
# Create gtab instance from bvals and bvecs
bvals = np.array([0., 10., 20., 30., 40., 60., 80., 100., 120., 140., 160., 180., 200.
                 , 250., 300.,350.])

N = len(bvals)
bvecs = get_bvecs(N) 
gtab = gradient_table(bvals, bvecs.T)

# Generate some data points taking arbitary parameters
S0= 100
f = 0.15*np.random.rand(5)
D = 0.00018*np.random.rand(5)
D_star = 0.028*np.random.rand(5)
```

```python
ivim_params = np.column_stack((f,D,D_star))
ivim_params
```

```python
predicted_signal = ivim_prediction(ivim_params, gtab)
plt.scatter(bvals, predicted_signal[0], )
plt.show()
```

```python
data = np.array(predicted_signal)
iv = IvimModel(gtab, fit_method="two_stage_fit")
fittedmodel = iv.fit(data, gtab)
```

```python
fittedmodel.plot(4)
```
